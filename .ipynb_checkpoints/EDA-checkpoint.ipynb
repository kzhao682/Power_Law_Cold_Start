## Group metadata

```python
for i in meta.columns[1:]:
    print(meta[i].value_counts())
```

Focus on days off first..the assumption is that the days off is a larger factor on the effect of energy consumption behavior.

```python
#encode True to 1 and False to 0
for i in meta.columns[3:]:
    meta.loc[meta[i]==True,i] = 1
    meta.loc[meta[i]==False,i] = 0
    
#create column with days off array
meta['days_off'] = [i for i in np.array(meta[['monday_is_day_off','tuesday_is_day_off','wednesday_is_day_off','thursday_is_day_off',
                                 'friday_is_day_off','saturday_is_day_off','sunday_is_day_off']])]
meta['days_off'] = meta['days_off'].apply(lambda x: ''.join(map(str, x)))
```
